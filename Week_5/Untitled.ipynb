```python
def to_bin(bits):
    hex_to_bin = {'0' : "0000",
                '1' : "0001",
                '2' : "0010",
                '3' : "0011",
                '4' : "0100",
                '5' : "0101",
                '6' : "0110",
                '7' : "0111",
                '8' : "1000",
                '9' : "1001",
                'A' : "1010",
                'B' : "1011",
                'C' : "1100",
                'D' : "1101",
                'E' : "1110",
                'F' : "1111" }
    binary_format = ""
    for hex_digit in bits:
        binary_format += hex_to_bin[hex_digit]
    return binary_format

def to_hex(bits):
    bin_to_hex = {"0000" : '0',
                "0001" : '1',
                "0010" : '2',
                "0011" : '3',
                "0100" : '4',
                "0101" : '5',
                "0110" : '6',
                "0111" : '7',
                "1000" : '8',
                "1001" : '9',
                "1010" : 'A',
                "1011" : 'B',
                "1100" : 'C',
                "1101" : 'D',
                "1110" : 'E',
                "1111" : 'F' }
    hex_format = ""
    for i in range(0,len(bits),4):
        binary_digits = bits[i:i+4]
        hex_format = hex_format + bin_to_hex[binary_digits]
    return hex_format
```

```python
def hexbyte_xor(A, B):
    byte1 = to_bin(A)
    byte2 = to_bin(B)
    result = ""
    for bit1, bit2 in zip(byte1, byte2):
        if bit1 == bit2:
            result += ("0")
        else:
            result += ("1")
    return to_hex(result)
```
